# Research Question 1: Gender Differences in Body Self-Image Scores

## Question: Do men and women differ significantly in body self-image perceptions, specifically in height dissatisfaction and fatness evaluation scores?

## Variables:

### DEMO_gender: This variable categorizes participants by gender, allowing for comparisons in body self-image across groups.

### PSYCH_body_self_image_questionnaire_height_dissatisfaction_score and  PSYCH_body_self_image_questionnaire_fatness_evaluation_score: Measures of participants' dissatisfaction with their height and body fatness, respectively.

## Method of Analysis: Independent t-tests to compare self-image scores between genders.

## Procedure:

### Check Normality (Visual Check): Plot a histogram of self-image scores for each gender. If the distribution looks roughly bell-shaped, the data is likely close enough to normal. Alternatively, use a boxplot to check for symmetry. If the boxplot is roughly symmetrical, it suggests the data may be approximately normal.

### Test Homogeneity of Variances: Use Levene’s test to check if variances are equal across gender groups. If not, use Welch’s t-test, which doesn’t assume equal variances.

### Perform T-test: Run the t-test (or Welch’s t-test if variances are unequal) to see if there are statistically significant differences between gender groups.

## Assumptions:

### Normality of self-image scores within each gender group (tested via Shapiro-Wilk).

### Independence of observations.

### Homogeneity of Variances across gender groups (tested with Levene’s test).

## Visualizations:

### Histogram for gender distribution to illustrate the sample’s gender composition.

### Boxplot of self-image scores by gender to visualize central tendencies and variability across genders.

## Hypothesis:

### Research Hypothesis: Women will report higher dissatisfaction scores due to societal pressures on body image.

### Null Hypothesis (H0): There is no significant difference in body self-image scores between men and women.

### Expected Results: If the research hypothesis holds, women will show significantly higher dissatisfaction scores than men, aligning with body image literature. This will answer whether gender impacts self-image in this sample.

# Research Question 2: Age Group Differences in Big Five Personality Traits

## Question: How do neuroticism and openness scores differ across age groups?

## Variables:

### DEMO_age: Represents participant age, allowing for categorization into age groups.

### PSYCH_big_five_inventory_neurotic_score and PSYCH_big_five_inventory_open_score: Scores on neuroticism and openness, two dimensions of the Big Five personality traits.

## Method of Analysis: Analysis of Variance (ANOVA) to test for differences in personality traits across age groups.

## Procedure:

### Check Normality (Visual Check):

### Plot histograms for neuroticism and openness scores within each age group to visually inspect for bell-shaped distributions.

### Use boxplots to assess the symmetry of scores across age groups; symmetrical boxplots suggest approximate normality.

### Test Homogeneity of Variances: Use Levene’s test to confirm equal variances across age groups. If unequal, apply Welch’s ANOVA.

### Conduct ANOVA: Run the ANOVA to test for significant differences between age groups in neuroticism and openness scores.

## Assumptions:

### Normality of personality trait scores within each age group.

### Independence of observations.

### Homogeneity of Variances across age groups.

## Visualizations:

### Histogram for age distribution to show age demographics.

### Boxplot of neuroticism and openness scores by age group to highlight potential differences.

## Hypothesis:

### Research Hypothesis: Younger participants will score higher in neuroticism, while older participants will score higher in openness, reflecting typical personality development trends.

### Null Hypothesis (H0): There is no significant difference in neuroticism and openness scores across age groups.

### Expected Results: Results are expected to reveal higher neuroticism among younger participants and higher openness among older ones, contributing insights into age-related personality traits.


# Research Question 3: Correlation Between Social Dependence and Extraversion

## Question: Is there a positive correlation between social dependence in body self-image and extraversion?

## Variables:

### PSYCH_body_self_image_questionnaire_social_dependence_score: Measures participants’ reliance on social approval for self-image.

### PSYCH_big_five_inventory_extraverted_score: Reflects the extraversion trait, related to social engagement.

## Method of Analysis: Pearson correlation to assess the relationship between social dependence and extraversion.

## Procedure:

### Assess Linearity (Visual Check):

### Create a scatterplot of social dependence vs. extraversion. If the points roughly follow a straight line, it indicates a linear relationship suitable for Pearson correlation.

### Check Normality (Visual Check): Plot histograms for both social dependence and extraversion scores. If they look roughly bell-shaped, they may be close enough to normal for the correlation test.

##Calculate Pearson Correlation: Compute the correlation coefficient to measure the strength and direction of the relationship.

## Assumptions:

### Linearity in the relationship (checked via scatterplot).

### Normality of both variables.

### Independence of observations.

## Visualizations:

### Scatterplot to visually assess the correlation between social dependence and extraversion.

### Histogram of extraversion scores to show distribution.

## Hypothesis:

### Research Hypothesis: There will be a positive correlation between social dependence and extraversion, as more extraverted individuals are hypothesized to rely more on social validation.

### Null Hypothesis (H0): There is no significant correlation between social dependence and extraversion.

### Expected Results: A significant positive correlation would suggest that higher extraversion aligns with greater social dependence, answering if social personality traits affect self-image reliance.
