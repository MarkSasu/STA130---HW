## What is the difference between a null hypothesis and an alternative hypothesis in the context of hypothesis testing?

- Null hypothesis: A claim that there is no difference or impact between variables or groups is known as the null hypothesis. This is the presumption that serves as the basis for comparing your data. Instead of conclusively proving or disproving the null hypothesis, the goal of hypothesis testing is to present evidence against it. In order to apply the scientific method to find evidence against a null hypothesis and proceed, we frequently establish one that we may not believe.

- Alternative Hypothesis (HA): The null hypothesis is simply refuted by the alternative hypothesis. It asserts that the null hypothesis is incorrect, suggesting that variables or groups differ or have an effect. Since it implies that a genuine occurrence is happening, this is sometimes the more "interesting" theory.

- An example from the tutorial: the null hypothesis in the "Stella's Wheel of Destiny" example is that the wheel is fair, with a 50/50 chance of landing on either black or red. The other possibility is that the wheel is unfair. A p-value can be computed using the gathered data (multiple spin results). If the wheel were truly fair, a low p-value would indicate that the observed results are implausible. As a result, the null hypothesis would be rejected, and it would be determined that the wheel might be biased. It's important to keep in mind, though, that this only offers evidence against the presumption of fairness and does not prove the wheel is biased.

## 3. The second "Pre-lecture" video (above) explains that we "imagine a world where the null hypothesis is true" when calculating a p-value? Explain why this is in your own words in a way that makes the most sense to you.

We "imagine a world where the null hypothesis is true" while calculating a p-value because we're estimating the likelihood that our data—or something more extreme—will be observed if the null hypothesis is true. To find out what kinds of outcomes we would anticipate by chance, we use the sampling distribution of the test statistic under the null hypothesis. We have cause to question the null hypothesis if our observed outcome is extremely unlikely to occur within this range. This likelihood is expressed by the p-value, where a low p-value indicates that the data is surprising under the null and a high p-value indicates that the data is consistent with the null. By using this method, we may assess how anomalous our finding would be if the null hypothesis were correct.

The null hypothesis appears more "ridiculous" when the p-value is less because it implies that the observed result would be incredibly implausible if the null hypothesis were true. Assuming the null hypothesis is true, the p-value is the likelihood of obtaining a test statistic that is as extreme as or more extreme than the one that was observed. The sampling distribution of the test statistic under the null hypothesis, which displays every value we may anticipate observing through repeated sampling if the null were true, can be used to comprehend this. Our observed result falls well out in the "tails" of this distribution, in a range that would be extremely unlikely if the null hypothesis were true, when the p-value is very small. This implies that under the null hypothesis, the data we gathered is highly improbable to have happened by accident. The p-value is smaller for more severe results because they are less likely to occur. It gets more difficult to accept that the null hypothesis offers a plausible explanation for the facts as the p-value decreases. Put another way, the more "ridiculous" the null hypothesis appears, the smaller the p-value, since it is less likely that the observed outcome could occur if the null hypothesis were true. Because the null hypothesis no longer looks like a likely explanation for the data we observed, a very small p-value frequently causes us to reject it in favor of the alternative.

## 5. Güntürkün (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a p-value using a "50/50 coin-flipping" model for the assumption of the null hypothesis that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against 

### Problem setup:
Null hypothesis: The probability of tilting heads either left or right is 50% (no preference).
Alternative hypothesis: The probability of tilting heads to the right is different from 50%.
Data: 80 out of 124 couples tilted their heads to the right.
Sample proportion: 80/124 = 0.645 = 64.5%
We now want to calculate the p-value based on this observed proportion, under the assumption that the true proportion is 50% (the null hypothesis).

### Steps to calculate the p-value:
1. Assume the null hypothesis is true: The expected proportion of head tilts to the right is 0.50.
2. Calculate the test statistic: This involves determining how far our observed proportion (64.5%) is from the expected 50% under the null hypothesis.
3. Simulate or calculate the probability of observing a sample proportion as extreme as 0.645, assuming the null hypothesis is true. This will give us the p-value.

```python
import scipy.stats as stats

# Given values
observed_successes = 80
total_couples = 124
null_hypothesis_proportion = 0.50

# Sample proportion
observed_proportion = observed_successes / total_couples

# Standard error for a proportion
std_error = (null_hypothesis_proportion * (1 - null_hypothesis_proportion) / total_couples) ** 0.5

# Z-score for the observed proportion
z_score = (observed_proportion - null_hypothesis_proportion) / std_error

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

p_value
```

## Problem Introduction:
The experiment revolves around replicating the famous "Fisher’s Tea Experiment," where Dr. Muriel Bristol claimed to taste the difference between cups of tea depending on whether milk was poured before or after the tea. In this experiment, we test a random sample of 80 STA130 students to see if they can correctly identify whether tea or milk was poured first, just like Dr. Bristol. Out of the 80 students, 49 were able to correctly state which was poured first.

We aim to perform a statistical analysis to determine whether the students are merely guessing (null hypothesis) or if there is some skill involved in identifying the correct pouring order (alternative hypothesis). This analysis includes setting up hypotheses, performing a hypothesis test, and interpreting the p-value based on the observed data.

## Relationship to the Original Fisher and Bristol Experiment:
In the original Fisher experiment, the sample size was smaller (only 8 cups of tea), and the parameter in question was Dr. Bristol’s ability to distinguish between cups of tea with differing pouring orders. The sample for our experiment is different in size and nature. Instead of testing one individual, we’re testing a group of students, and the parameter of interest is whether these students, on average, can correctly distinguish the pouring order or are simply guessing.

## Null and Alternative Hypotheses:
Null Hypothesis: The students are randomly guessing, which means they have a 50% probability of correctly identifying whether the tea or milk was poured first. Mathematically, this translates to the population proportion being  p=0.5.

Alternative Hypothesis: The students are not randomly guessing and have a better-than-random ability to identify the correct pouring order. Mathematically, the population proportion p>0.5.

Informal Statement of the Hypotheses:
    Null Hypothesis: "The students are just guessing and are correct about half the time."
    Alternative Hypothesis: "The students are better than random at correctly identifying which was poured first."

## Quantitative Analysis:
Test Statistic: We will use the sample proportion of correct identifications, which is 0.6125

Method: To assess the validity of the null hypothesis, we will perform a hypothesis test using a binomial test or a normal approximation of the binomial distribution. Given the sample size of 80, we can approximate the binomial distribution using the normal distribution to calculate the p-value for the observed proportion.

Purpose of the Statistic: The test statistic allows us to measure the distance between the observed sample proportion (0.6125) and the hypothesized population proportion (0.5) under the null hypothesis. This will allow us to assess whether the observed results are significantly different from what we’d expect by random guessing.

## Code and methodology

```python
import numpy as np
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(130)

# Sample data
n = 80  # number of students
successes = 49  # number of correct identifications
p_null = 0.5  # null hypothesis: random guessing, 50% chance

# Sample proportion
p_hat = successes / n

# Binomial test (exact) using 'binomtest'
p_value_binom = stats.binomtest(successes, n=n, p=p_null, alternative='greater')
print("Binomial Test P-Value:", p_value_binom.pvalue)

# Normal approximation
standard_error = np.sqrt(p_null * (1 - p_null) / n)
z_score = (p_hat - p_null) / standard_error
p_value_normal = 1 - stats.norm.cdf(z_score)  # One-sided test
print("Normal Approximation P-Value:", p_value_normal)

# Confidence Interval using normal approximation (95%)
z_critical = stats.norm.ppf(0.975)  # Two-tailed, 95% CI
ci_lower = p_hat - z_critical * standard_error
ci_upper = p_hat + z_critical * standard_error
print(f"95% Confidence Interval for the true proportion: ({ci_lower:.3f}, {ci_upper:.3f})")
```

## Findings and Discussion:
P-value (Binomial Test): The p-value from the binomial test is
p≈0.022. This suggests that if the null hypothesis were true, the probability of observing 49 or more correct identifications (or something more extreme) by random guessing is about 2.2%. P-value (Normal Approximation): The p-value using the normal approximation is similar, approximately p≈0.021, reinforcing the conclusion. Confidence Interval: The 95% confidence interval for the true proportion of correct identifications is approximately 
(0.503,0.722). Since 0.5 is just within the lower bound of the interval, this suggests that while the results lean toward students having a better-than-random ability, the evidence is not overwhelmingly strong.

## Conclusion Regarding the Null Hypothesis:
Given the p-value is below the common significance level of 0.05, we reject the null hypothesis. This means there is moderate evidence that the STA130 students are better than random at identifying whether milk or tea was poured first. However, the evidence is not extremely strong, and further tests or a larger sample size might be necessary to strengthen this conclusion.
