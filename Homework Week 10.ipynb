### 1.  A simple Linear regression is a method that medols the relationship between a single dependant variable (outcome) and the independant variable (predictor), where Y represents the dependant variable. X is the independant variable, B0 is the intercept and B1 is the slope of the line while ϵ represents the error term. You may notice that this is similar to the equation of a tangent line, y = mx + b. On the contrary,  multiple linear regression is an extension of the simple linear regression that models the relationship in between multiple dependant and independant variables. A benefit of working with Multiple Linear Regression over Simple Linear Regression is Realistic Modeling. The benefit is as simple as it seems; real-world results are often influenced by a variety of factors, which is why Multiple linear regression is better equiped due to the fact that is is more applicable in complex analyses.

### 2. There are several difference that can be accounted for between the two linear forms. Firstly, the nature of the relationship, the difference in between the two is that a continuous variable represents a gradient relationship, where the change in the outcome variable is proportional to the change in the input variable. In an indicator variable, the relationship between the outcome and input variable represents a shift in the mean value of the outcome variable between two categories, not exactly a continuous change. Another difference is the graphical representation, a continous variable results in a continous line graph that is uninterupted mean while for the indicator variable, the coefficient B1 shows the difference between the different Y va;ues from the two groups represented by the indicator variable.

### 3. The simple linear regression provides a single intepretation of how the dependant variable differentiates as the independant variables changes. There are no outside factors or specific effects that are being considered when using simple linear regression. Therefore, any factor influencing Y must be only shown in X alone. For a multiple linear regression with continuous and indicator variables, when an indicator, let's say D is introduced. The model is able to differentiate different subgroups of the model and return knowledge on how Y is infleunced across these different groups. B1 still represents the effect of the continuous variable X on Y, but now the effect is considered within the context of specific categories defined by D. B2 represents the difference in intercepts between the two categories defined by D. For instance, if D represents gender, B2 quantifies how the average salary differs between genders when controlling for years of experience. You may ask: What is the benefit of moving from SLR to MLR. Well, the incllusion of multiple predictors; what this means is the MLR allows for a more comprehensive modeling by incorporating muktple predictors, enabling for a more accurate and overall explanations of the dependant variable (Y).

### 4. Adding an interaction between a continuous and an indicator variable in a Multiple Linear Regression Model allows the model to capture how the effect of the continuous variable on the dependant variable Y changes depending on the cateogry defined by the indicator. This is because adding an interaction term makes the model more flexible and capable of modeling complex relationships. There are several key points of adding an interaction term though, firstly, there are different slpos. If the interaction term is significant, it indicates that the effect of x on y changes between categories. Furthermore, it provides enhanced insight, it represents a deeper understanding of how different factors work together to affect the outcome, making the model better at representing real-world scenarios.

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```
